```python
# 导入库
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 天气预测train
train = pd.read_csv("./data/train.csv")
```

```python
# 折线图
x = train["id"]
y = train["Rainfall"]

plt.title('weather')  # 设置标题
plt.xlabel('id')  # 设置 x 轴标签
plt.ylabel('Temp9am')  # 设置 y 轴标签
plt.grid(True)  # 显示网格

plt.plot(x, y,linestyle = 'dotted',marker = "None")
plt.scatter(x, y)

plt.show()
```
